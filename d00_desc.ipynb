```python
def time_metric_down(time, metric, metric_down):
    
    if (metric == 'w' and metric_down == 'd'):
        x = 7
    elif (metric == 'w' and metric_down == 'h'):         
        x = 7 * 24
    elif (metric == 'w' and metric_down == 'm'):         
        x = 7 * 24 * 60
    elif (metric == 'w' and metric_down == 's'):         
        x = 7 * 24 * 60 * 60
    elif (metric == 'd' and metric_down == 'h'):         
        x = 24
    elif (metric == 'd' and metric_down == 'm'):         
        x = 24 * 60
    elif (metric == 'd' and metric_down == 's'):         
        x = 24 * 60 * 60  
    elif (metric == 'h' and metric_down == 'm'):         
        x = 60
    elif (metric == 'h' and metric_down == 's'):         
        x = 60 * 60  
    elif (metric == 'm' and metric_down == 's'):         
        x = 60 
    return time * x  

if (time_metric_down(10, 'h', 's') == 36000 and time_metric_down(1, 'w', 'd') == 7 and 
    time_metric_down(1, 'w', 'h') == 168 and time_metric_down(1, 'w', 'm') == 10080 and 
    time_metric_down(1, 'w', 's') == 604800):
    print('Ошибки не выявлены')
else:
    print('Выявлены ошибки')
```

```python
from timeit import timeit
import numpy as np
import math
import sys
sys.setrecursionlimit (1500)

print('OK')

def factorial1(n):
    fact = 1
    for i in range(2, n + 1):
        fact *= i
    return fact

def factorial2(n):
    fact = 1
    while n > 1:
      fact *= n
      n -= 1
    return fact

def factorial3(n):
    if n == 0:
        return 1
    return factorial3(n-1) * n

def factorial4(n):
    return math.factorial(n)

func = ['factorial1', 'factorial2', 'factorial3', 'factorial4']
time = []

if __name__ == '__main__':
  for j in func:
    time.append(timeit(j+'(8)', setup = 'from __main__ import '+j, number = 100))

time = np.array(time)                              
time = time*100/time.max()  

for i, j in zip(func, time):
    print(i, '-', str(round(j,2))+'%')

def best_fctrl(n):
    return math.factorial(n)

if (best_fctrl(0) == 1 and best_fctrl(1) == 1 and best_fctrl(2) == 2 and best_fctrl(3) == 6 and 
    best_fctrl(10) == 3628800):
    print('Ошибки не выявлены')
else:
    print('Выявлены ошибки')
```

```python
import math

workers = int(input("Введите количество рабочих (целое положительное число) = "))
tasks = int(input("Введите количество задач (целое положительное число) = "))
time = int(input("Введите время на одну задачу (целое положительное число) = "))
while(1):
  overal_time = int(input("Введите время, за которое необходимо уложиться (целое положительное число) = "))
  if (overal_time < time):
    print('Невозможно выполнить задачу в этих условиях, введите новое время')
  else:
    break
  
def new_workers(workers, tasks, time, overal_time):
  n_session = overal_time // time
  new_workers = math.ceil(tasks / n_session) - workers
  if new_workers < 0:
    new_workers = 0
  return new_workers

n = new_workers(workers, tasks, time, overal_time)
print('Необходимо нанять ', n, 'новых работников')
```
